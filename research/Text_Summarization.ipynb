```python
from transformers import pipeline,set_seed
from transformers import AutoTokenizer, AutoModelForCausalLM
from datasets import load_dataset,load_from_disk
import torch
import matplotlib.pyplot as plt
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize
nltk.download("punkt")
```

```python
device="cuda" if torch.cuda.is_available() else "cpu"
model_ckbt="google/pegasus-cnn_dailymail"
tokenizer=AutoTokenizer.from_pretrained(model_ckbt)
model=AutoModelForCausalLM.from_pretrained(model_ckbt).to(device)
```

```python
dataset_samsum=load_from_disk("samsum_dataset")
dataset_samsum
```

```python
split_lengths=[len(dataset_samsum[split]) for split in dataset_samsum]
print(f"Split lengths: {split_lengths}")
print(f"Features: {dataset_samsum['train'].column_names}")
print("\nDialogue:")
print(dataset_samsum["test"][0]["dialogue"])
print("\nSummary:")
print(dataset_samsum["test"][0]["summary"])
#
```

```python
def convert_examples_to_features(example_batch):
  input_encodings=tokenizer(example_batch["dialogue"],max_length=1024,truncation=True,padding="max_length")
  with tokenizer.as_target_tokenizer():
    target_encodings=tokenizer(example_batch["summary"],max_length=1024,truncation=True,padding="max_length")
  return {
      "input_ids":input_encodings["input_ids"],
      "attention_mask":input_encodings["attention_mask"],
      "labels":target_encodings["input_ids"]
  }
```

```python
dataset_samsum_pt=dataset_samsum.map(convert_examples_to_features,batched=True)
```

```python
dataset_samsum_pt["train"]
```

```python
from transformers import DataCollatorForSeq2Seq
seq2seq_data_collator=DataCollatorForSeq2Seq(tokenizer,model=model)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="pegasus-samsum",
    num_train_epochs=1,
    warmup_steps=500,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    weight_decay=0.01,
    logging_steps=10,
    # Replace evaluation_strategy with the appropriate argument if needed
    # For instance, if you want to evaluate at the end of each epoch:
    #evaluation_strategy = "epoch",
    # Or if you want to evaluate every 500 steps:
    eval_steps=500,
    save_steps=1e6,
    gradient_accumulation_steps=16
)
```

```python
trainer=Trainer(model=model,args=training_args,
                tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                train_dataset=dataset_samsum_pt["test"],eval_dataset=dataset_samsum_pt["validation"])
```

```python
trainer.train()
```

```python
import nltk
import numpy as np
from rouge_score import rouge_scorer

def evaluate_summaries(predictions, references):
    """
    Evaluates summaries using ROUGE scores.

    Args:
        predictions: A list of generated summaries.
        references: A list of reference summaries.

    Returns:
        A dictionary containing the average ROUGE scores.
    """
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = []
    for ref, pred in zip(references, predictions):
        score = scorer.score(ref, pred)
        scores.append(score)
    avg_scores = {
        'rouge1': np.mean([s['rouge1'].fmeasure for s in scores]),
        'rouge2': np.mean([s['rouge2'].fmeasure for s in scores]),
        'rougeL': np.mean([s['rougeL'].fmeasure for s in scores]),
    }
    return avg_scores
```

```python
from tqdm import tqdm

def generate_summaries(dataset, start_index=0, end_index=10):
    """
    Generates summaries for a subset of the given dataset.

    Args:
        dataset: The dataset to generate summaries for.
        start_index: The starting index of the subset.
        end_index: The ending index of the subset.

    Returns:
        A tuple containing the generated summaries and reference summaries.
    """
    predictions = []
    references = []

    for i in tqdm(range(start_index, end_index)):  # Iterate through the specified range
        example = dataset[i]  # Get the example at the current index
        inputs = tokenizer(example["dialogue"], return_tensors="pt").to(device)
        generated_ids = model.generate(**inputs, max_length=1024)
        generated_summary = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
        predictions.append(generated_summary)
        references.append(example["summary"])

    return predictions, references

predictions, references = generate_summaries(dataset_samsum_pt["validation"], start_index=0, end_index=11)  # Generate for indices 0 to 10
```

```python
results = evaluate_summaries(predictions, references)
print(results)
```
